## Data Cleaning
Remove Unnecessary Columns: The Unnamed: 0 column seems redundant and can be removed.
Handle JSON-like Columns: The columns genres, keywords, cast, and crew are in a JSON-like format and need to be processed into a more usable form.
Handling Missing Values: Check for and handle missing values in the dataset.
Convert Data Types: Ensure that columns like release_date are in the correct format.

```python
# Data Cleaning Process

# 1. Remove Unnecessary Column
data_cleaned = data.drop(columns=['Unnamed: 0'])
```

```python
# 2. Handle JSON-like Columns
# For simplicity, we'll just extract the first genre, keyword, cast, and crew member for each movie
import json

def extract_first_item(json_str):
    try:
        json_data = json.loads(json_str.replace("'", "\""))
        if len(json_data) > 0:
            return json_data[0]['name']
        return None
    except json.decoder.JSONDecodeError:
        return None

data_cleaned['first_genre'] = data_cleaned['genres'].apply(extract_first_item)
data_cleaned['first_keyword'] = data_cleaned['keywords'].apply(extract_first_item)
data_cleaned['first_cast'] = data_cleaned['cast'].apply(extract_first_item)
data_cleaned['first_crew'] = data_cleaned['crew'].apply(extract_first_item)
```

```python

# 3. Handling Missing Values
missing_values = data_cleaned.isnull().sum()

# 4. Convert Data Types
data_cleaned['release_date'] = pd.to_datetime(data_cleaned['release_date'], errors='coerce')

# Display the changes and missing values
data_cleaned.head(), missing_values.sort_values(ascending=False)
```

## Feature Engineering

Creating binary or categorical variables for top genres.

Generating time-related features from the release date, such as the movie's age or the release quarter.
Handling missing values, especially in the first_cast, tagline, first_crew, and first_keyword columns.

```python
# Feature Engineering

# 1. Creating binary variables for top genres (we'll choose the top 5 for simplicity)
top_genres = data_cleaned['first_genre'].value_counts().index[:5]
for genre in top_genres:
    data_cleaned[f'genre_{genre}'] = data_cleaned['first_genre'].apply(lambda x: 1 if x == genre else 0)

# 2. Encoding the original language as binary features (focusing on non-English movies)
top_languages = data_cleaned['original_language'].value_counts().index[1:6] # Excluding English
for language in top_languages:
    data_cleaned[f'language_{language}'] = data_cleaned['original_language'].apply(lambda x: 1 if x == language else 0)

# 3. Generating time-related features
data_cleaned['movie_age'] = 2023 - data_cleaned['release_year']  # Assuming the current year is 2023
data_cleaned['release_quarter'] = data_cleaned['release_date'].dt.quarter

# 4. Handling missing values
# For simplicity, we'll fill missing values with a placeholder for now
data_cleaned = data_cleaned.fillna('Unknown')

# Displaying the changes
data_cleaned.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Feature Selection
# We're selecting 'first_genre', 'first_keyword', 'first_cast', 'first_crew', and 'overview' for content-based filtering
features = ['first_genre', 'first_keyword', 'first_cast', 'first_crew', 'overview']

# Text Processing: Overview
# Using TF-IDF to convert text data to a usable format
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data_cleaned['overview'])

# We will only use overview for the sake of simplicity and demonstration
# In a real-world scenario, you might want to combine multiple features

# Similarity Scoring: Using Cosine Similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get recommendations based on similarity scores
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = data_cleaned[data_cleaned['original_title'] == title].index[0]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return data_cleaned['original_title'].iloc[movie_indices]

# Testing with an example movie
example_movie = "Avatar"
recommendations = get_recommendations(example_movie)
recommendations
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re
from nltk.stem import PorterStemmer

# Initialize Stemmer
stemmer = PorterStemmer()

def preprocess_text_without_stopwords(text):
    # Lowercase conversion
    text = text.lower()
    # Removing special characters and digits
    text = re.sub(r'\W+|\d+', ' ', text)
    # Tokenization (splitting text into words)
    words = text.split()
    # Stemming (without stopword removal)
    words = [stemmer.stem(word) for word in words]
    # Joining back to form processed text
    return ' '.join(words)

# Applying preprocessing to overview column without stopwords
data_cleaned['processed_overview'] = data_cleaned['overview'].apply(preprocess_text_without_stopwords)
```

```python
# Text Processing: Processed Overview
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(data_cleaned['processed_overview'])

# Similarity Scoring
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim_matrix, data):
    # Get the index of the movie that matches the title
    idx = data[data['original_title'] == title].index[0]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim_matrix[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return data['original_title'].iloc[movie_indices]

# Using the function with the computed cosine similarity matrix
recommendations = get_recommendations("Avatar", cosine_sim, data_cleaned)
print(recommendations)
```
